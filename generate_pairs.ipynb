```python
with open('common_words.txt') as f: 
    words = [w.strip() for w in f.readlines()]
words[0]
```

```python
words_by_length = {}
for word in words: 
    if not words_by_length.get(len(word)):
        words_by_length[len(word)] = [word]
    else: 
        words_by_length[len(word)].append(word)
words_by_length[2]
```

```python
from collections import Counter
import itertools

combine_words = words_by_length[3] + words_by_length[4] + words_by_length[5]
len(combine_words)
```

```python
from tqdm import tqdm

matches = []
pbar = tqdm(enumerate(combine_words), total=len(combine_words))
for i, word1 in pbar:
    for word2 in combine_words[i:]:
        combined = word1 + word2
        counted = Counter(combined)
        for scramble in words_by_length[len(combined)]:
            if Counter(scramble) == counted:
                matches.append((word1, word2, scramble))
                pbar.set_description(f'total matches = {len(matches)}')
```

```python
matches[:5]
# Good anagrams
unique_matches = [m for m in matches if m[0] not in m[2] and m[1] not in m[2]]
unique_matches[:5]
```

```python
import json 
with open("data.json", "w") as f:
    json.dump(unique_matches, f)
```
